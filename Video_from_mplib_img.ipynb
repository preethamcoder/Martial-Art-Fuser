```python
import os
from PIL import Image
import imageio

def split_image(file_path, out_folder, num_images):
    imgs = os.listdir(file_path)
    for image in os.listdir(file_path)[1:]:
        image = file_path+"/"+image
        img = Image.open(image)
        img_width, img_height = img.size
        slice_width = img_width // num_images

        if not os.path.exists(out_folder):
            os.makedirs(out_folder)

        for i in range(num_images):
            left = i * slice_width
            right = (i + 1) * slice_width
            slice_img = img.crop((left, 0, right, img_height))
            slice_img.save(os.path.join(out_folder, f'frame_{i}.png'))

def create_video(image_folder, video_name, fps=30):
    image_files = sorted([img for img in os.listdir(image_folder) if img.endswith(".png")])

    with imageio.get_writer(video_name, fps=fps) as writer:
        for image_name in image_files:
            image_path = os.path.join(image_folder, image_name)
            image = imageio.imread(image_path)
            writer.append_data(image)

# Path to your .png file with multiple images
file_path = 'fin/'
out_folder = 'extracted_frames/'
num_images = 32  # Number of images in the .png file

split_image(file_path, out_folder, num_images)

# Create a video from the extracted frames
video_name = 'video_from_frames.mp4'
create_video(out_folder, video_name)
```
